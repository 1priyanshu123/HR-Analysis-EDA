```python
# What is the total absences and average engagement survey for each departments
df.groupby('Department').agg({'Absences':'sum','EngagementSurvey':'mean'})
```

```python
# What is the maximum salary and minimum late in the last 30 days for employee in each position?

df.groupby('Employee_Name').agg({'Salary':'max','DaysLateLast30':'min'})
```

```python
# How many Terminated emp were there in each department and what is the average empsatisfaction level among them?

df.groupby('Department').agg({'Termd':'sum','EmpSatisfaction':'mean'})
```

```python
#what is the earliesy and latest date of the hire for employees in each manager's team ? : 
earliest_dateof_hire=df.groupby('ManagerName').agg({'DateofHire':min}).reset_index() 
latest_dateof_hire=df.groupby('ManagerName').agg({'DateofHire':max}).reset_index() 
date_of_hiring=pd.merge(earliest_dateof_hire,latest_dateof_hire,on='ManagerName').rename(columns={'DateofHire_x':"Earliest_DateofHire",'DateofHire_y':"Latest_DateofHire"}) 
date_of_hiring
```

```python
# What is the total no of special projects and average absences for emp in each gender category?

df.groupby('Sex').agg({'SpecialProjectsCount':'sum','Absences':'mean'})
```
